```python
import os
import matplotlib.pyplot as plt
from datetime import datetime
import numpy as np
import time
import word2vec
import string
import pickle
import nltk
from nltk.stem.wordnet import WordNetLemmatizer
from nltk.stem.porter import PorterStemmer
from nltk.corpus import reuters
from nltk import word_tokenize
import operator
import json
import gensim
import re

from  pycorenlp  import StanfordCoreNLP
nlp = StanfordCoreNLP('http://localhost:9000')
```

```python
def read_file(csv):
    """
    Read csv file and return a list.
    """
    with open(csv, 'r', encoding='utf-8') as f:
        return f.read().split('\n')

def process_data(file):
    #Remove punctuations and headlines with UPDATE-tag and other tags
    
    article = []
    for line in file:
        data = line.split(',')
        if len(data) < 4:
            continue
        article += [(data[0], data[2], data[3])]
    
    titles = []
    for t in article:
        title = t[2].strip()
        title = re.replace('  ', ' ')
        title = re.sub(r'\'', ' ', title)
        title = re.sub(r'\$', ' $ ', title)
        title = re.sub(r' yr ', ' year ', title)
        for i in range(3):
            if '-' in title and (title[title.find('-') - 1].isupper() or 'UPDATE' in title):
                title = title[title.find('-') + 1:]
            else:
                break
        
        translator = str.maketrans('', '', string.punctuation)
        title = title.translate(translator)
        title = title.replace('  ', ' ')
        if len(titles) > 0 and title in titles[-1][2]:
                continue
        titles += [(t[0], t[1], title)]
    return titles

def titles2head(titles):
    titles2headline = {}
    index = 0
    for t in titles:
        titles2headline[t[2]] = (titles[index])
        index += 1
    return titles2headline
        


def generate_phrases(titles, filein, fileout):
    text = '. '.join(t[2] for t in titles)
    with open(filein, 'w') as f:
        f.write(text)
    word2vec.word2phrase(filein, fileout, verbose=True)

def unify_word(word):  
    """unify verb tense and noun singular"""
    word = word.lower()
    # ADJ, ADJ_SAT, ADV, NOUN, VERB
    pos_tag = ['a', 's', 'r', 'n', 'v']
    for wt in pos_tag:
        try:
            word = WordNetLemmatizer().lemmatize(word, pos=wt)
        except:
            pass
    return word
```

```python
data = read_file('./input/news_reuters.csv')
titles = process_data(data)
titles2headline = titles2head(titles)

print(titles[:10])
```

```python
INPUT_TEXT = './input/texts/text.txt'
OUTPUT_PHRASES = './input/texts/text-phrases.txt'
generate_phrases(titles, INPUT_TEXT, OUTPUT_PHRASES)
```

```python
texts = ['']
index = 0
i = 0
with open('input/texts/text.txt', 'r') as f:
    for t in f.read().split('. '):
        if len(texts[index]) + len(t) > 1e4:
            index += 1
            texts += ['']
        texts[index] += t + ". "

print(len(texts))
```

```python
outputs = []
for i in range(len(texts)):
    output = nlp.annotate(texts[i], properties={
      'annotators': 'openie',
      'outputFormat': 'json'
      })
    #print((output['sentences'][0]['openie']))
    outputs += [output]
    #print(len(output['sentences']))


```

```python
data = {}
bad = 0
good = 0
repeat = 0
translator = str.maketrans('', '', string.punctuation)
for o in outputs:
    for s in o['sentences']:
        if len(s['openie']) > 0:
            try:
                info = titles2headline[' '.join([s['tokens'][i]['word'].translate(translator) 
                                            for i in range(len(s['tokens']))][:-1])]
    #             print s['openie']
                if info in data:
                    repeat += 1
                data[info] = s['openie']
                good += 1
            except:
                print(' '.join([s['tokens'][i]['word'] for i in range(len(s['tokens']))][:-1]))
                bad += 1
                print()
print(len(data), bad, good, repeat)

with open('openie.p', 'wb') as outfile:
    pickle.dump(data, outfile)
```

```python
from pywsd.utils import lemmatize_sentence
from nltk.corpus import stopwords
from gensim.models import FastText
stop_words = set(stopwords.words('english'))
from gensim.models import KeyedVectors
```

```python
index = 0
info2index = {}
index2info = []
word_emb = []
```

```python
for k in openie:
    info2index[k] = index
    index += 1
    index2info += [k]

    emb = np.zeros((3, 100))
    for e in openie[k]:
        sub = e['subject'].split(' ')
        for w in sub:
            if w not in word2idx:
                word2idx[w] = word_index
                word_index += 1
            if word2idx[w] not in We:
                We[word2idx[w]] = np.array(np.random.uniform(-1.0, 1.0, 100))
            emb[0] = We[word2idx[w]]
        rel = e['relation'].split(' ')
        for w in rel:
            if w not in word2idx:
                word2idx[w] = word_index
                word_index += 1
            if word2idx[w] not in We:
                We[word2idx[w]] = np.array(np.random.uniform(-1.0, 1.0, 100))
            emb[1] = We[word2idx[w]]
        obj = e['object'].split(' ')
        for w in obj:
            if w not in word2idx:
                word2idx[w] = word_index
                word_index += 1
            if word2idx[w] not in We:
                We[word2idx[w]] = np.array(np.random.uniform(-1.0, 1.0, 100))
            emb[2] = We[word2idx[w]]
    emb /= len(openie[k])
    word_emb += [emb]

with open('word_emb_info.p', 'wb') as outfile:
    pickle.dump({
        'info2index': info2index,
        'index2info': index2info
    }, outfile)

word_emb = np.array(word_emb)
print(word_emb.shape)
print(index, word_index)
print(len(info2index))
```
